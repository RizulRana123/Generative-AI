```python
#default argument-these argument value given during defined a function
def area_perimeter(width = 2, height = 4):
  area = width*height
  perimeter = 2*(width+height)
  return area, perimeter
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
    return sum
```

```python
print(sum_number(3, 4, 5, 6, 2))
```

```python
#keyword variable lenght argument
#keyword variable lenght argument behave as dictionary
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a = 10, b = 20, c = 30)
```

```python
def pass_students(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.item():
    if marks>33:
      pass_student_list.append(name)
      return pass_student_list
```

```python
pass_students(A= 40, B= 30, C=60)
```
