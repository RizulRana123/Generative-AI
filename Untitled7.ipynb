```python
cgpa = np.array([6, 6.5, 7, 7.8, 8, 8.4, 8.9, 9, 9.5, 10])
package = np.array([4, 4.5, 6, 6.5, 7, 7.8, 8, 8.9, 10, 12])
font1 = {"family":"serif", "color":"blue","size":20}
font2 = {"family":"serif", "color":"darkred","size":15}
plt.plot(cgpa, package,"*:g", ms = 15,mec = "red",mfc ="purple",linewidth = 5)
plt.xlabel("CGPA",fontdict=font2)
plt.ylabel("Package",fontdict=font2)
plt.title("CGPA vs Package",fontdict=font1)
plt.grid(color="pink",linestyle = ":",linewidth = 1.5)
plt.show()
```

```python
import numpy as np
from numpy.linalg import norm

A = np. array([2, 1, 2, 3, 2, 9])
B = np.array([3, 4, 2, 4, 5, 5])

# compute cosine similarity
cosine = np.dot(A,B)/(norm(A)*norm(B))
print("cosine similarity" , cosine)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

text ="""The global shift towards renewable energy is accelerating,
driven by concerns over climate change and energy security. solar
and wind power installations have reached record levels worldwide."""

summary = summarizer(text, max_length=30, min_length=10, do_sample=False)
print(summary[0]['summary_text'])
```
